```python
data = pd.read_csv('Mountains.csv')
ParMount = data['Parent mountain']
counts = {}
```

```python
for key in ParMount:
    counts[key]=0
   
```

```python
for index in ParMount:
    for key, value in counts.items():
        if index == key:
            counts[key] += 1      
```

```python
total_result = {}
for key, value in counts.items():
    if value > 3:
       total_result[key] = value
```
